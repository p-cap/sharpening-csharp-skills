### array vs collection

- to use an array, we need to assign a size and/or values to it
- when declaring an array with a size, we can add values by referring to its index

```
int[] mynums = new int[5];

mynums[0] = 7;
mynums[4] = 67;

for (int i = 0; i < mynums.Count(); i++)
{
   Console.WriteLine($"index: {i} - {mynums[i]}");
}
```

- using `collections` allow me to add and remove elements dynamically
- use `Add` instead of `Append`

```
List<string> mystring = [];
mystring.Add("wassup");

foreach (string item in mystring)
{
    Console.WriteLine(item);
}
```

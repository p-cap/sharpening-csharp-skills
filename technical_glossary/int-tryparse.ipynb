### `int.TryParse`

- call `int.TryParse` and pass in an int

- When using `int.TryParse()`, the parameter has to be a string

```
string one = "1";

int.TryParse(one, out int result);
Console.WriteLine(result);
```

- when passing an `int`, we get the `CS1503` error

```
int one = 1;
int.TryParse(one, out int result);
Console.WriteLine(result);
```

- when passing a `char`, we get the `CS1503` error

```
char two = '2';
int.TryParse(two, out int result);
Console.WriteLine(result);
```

null literal example

```
string name = null; // 'name' does not point to any string object
```

In C#, assigning a null value to a non-nullable type will result in a compile-time error. Non-nullable types are designed to always hold a value, thereby preventing null references that can lead to exceptions during runtime.

- Assigning a null value to a `non-nullable` type
- `int` type, as mentioned in the compile error, is a non-nullable value type

```
int one = null;
```

- `int?` is considered a nullable type

```
int? meNullable = null;
Console.WriteLine(meNullable);
```

- adding a nullable type into the list 

```
string? title;
List<string> mylist = [];
mylist.Add(title);
```

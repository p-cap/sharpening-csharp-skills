### Bits and pieces of C# - Static keyword

- Using the `static` keyword in a method
- Why `static`?
    - you can call the Class name, not a variable -> this means no instance required
    - Cannot access instance data? -> a static function cannot use `this` or any variables (fields) in the class that aren't marked as static
    - Memory Efficient - because the function belongs to the type, only one copy exists in memory for the entire life of the program
- `Main` is the most famous static method

```
int sum = Calculator.Add(9,7);
Console.WriteLine($"The sum is: {sum}");
public class Calculator
{
    public static int Add(int a, int b) => a + b;
}
```

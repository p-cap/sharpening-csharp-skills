- Checks if a value of a specific type

- declared `Pcap` enum

```
public enum Pcap
{
    yo,
    wassup,
    howdy
}
```

- passing in the `yo` string to see if it is of type `Pcap`
- `Enum.IsDefined(typeof(Pcap), val)` returns True

```
string val = "yo";
Console.WriteLine(Enum.IsDefined(typeof(Pcap), val));
```

- lets pass a string not part of the `enum`

```
string val1 = "not_me";
Console.WriteLine(Enum.IsDefined(typeof(Pcap), val1));
```

`Enum.IsDefined(typeof(Pcap), val1)` will return bool

### C# enum demo

```
Console.WriteLine("#-- Yo Mundo --#");
Console.WriteLine("Enum Demo");
```

- declaring the enum

```
public enum TaskStatus
{
    Pending,
    InProgress,
    Completed
}
```

- printing the `enumuerator` from `enum`

```
Console.WriteLine($"TaskStatus.Pending: {TaskStatus.Pending}");
```

- iterate through the values of the `enum` instance

```
foreach (var status in Enum.GetValues(typeof(TaskStatus)))
{
    Console.WriteLine($"staus: {status}");
    Console.WriteLine($"(int)status: {(int)status}");
}
```

- print each int value of the `enum` individually

```
Console.WriteLine($"((int)TaskStatus.Completed): {((int)TaskStatus.Completed)}");
```

### Class demo

- declare the enum that will be a type for one of the task class attributes
- declare the class
- instantiate the class

- declaring a public `enum` 

```
public enum TaskStatus
{
    Pending,
    InProgress,
    Completed
}
```

- declare the `Task` class which is also public

```
public class Task
{
    // public attributes
    public string Title {get; set;}

    public TaskStatus Status {get; set;}

    public Task(string title)
    {
        Title = title;
        Status = TaskStatus.Pending;
    }
}
```

```
Task myTask = new Task("wassup");
Console.WriteLine($"myTask title: {myTask.Title}");
Console.WriteLine($"myTask status: {myTask.Status}");
```

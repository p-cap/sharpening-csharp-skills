#### Binary Search

- Generate a list test case
- set a target which is an element that can or cannot be found in in the list
- set the left and right pointers
- perform Binary search
- return True if the element exists otherwise False

```
using System.Globalization;
using System.Runtime;
using System.Runtime.InteropServices;

List<int> mynums = [];
bool result = false;

for (int i = 0; i < 10; i++)
{
    mynums.Add(i);
}

int target = 3;
int left = 0;
int right = mynums.Count - 1;

Console.WriteLine($"left: {left}");
Console.WriteLine($"right: {right}");

while (left <= right)
{
    int mid = (left + right + 1) / 2;
    if (mynums[mid] == target)
    {
        result = true;
        break;
    }

    else if (target > mynums[mid])
    {
        left = mid + 1;
    }

    else
    {
         right = mid - 1;
    }

}
Console.WriteLine(result);
```

Binary Search in my own words
- if the target is greater than mid
    - we eliminate the left side 
    - because the target is on the right of mid
    - we move the left pointer to the right of mid
    - `mid = left + 1`
- if the target is less than mid
    - we eliminate the right side
    - because the target is on the left side of mid
    - we move the right pointer to the left of mid
    - `mid = right - 1`

### Numbers in c sharp

- `int.MinValue` and `int.MaxValue`

```
int min = int.MinValue;
int max = int.MaxValue;
Console.WriteLine($"int.MinValue: {min}");
Console.WriteLine($"int.MaxValue: {max}");
```

- dividing doubles

```
double third = 1.0 / 3.0;
Console.WriteLine(third);
```

- dividing with int 

```
int anotherThird = 1 / 3;
Console.WriteLine(anotherThird);
```

- even though the answer is declared double, the answer is still an int

```
double test = 5 / 2;
Console.WriteLine(test);
```

- if one of the numbers is a double, it turns into a double

```
double test = 5 / 2.0;
Console.WriteLine(test);
```

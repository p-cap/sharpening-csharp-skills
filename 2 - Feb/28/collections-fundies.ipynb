- using the `List.IndexOf(....)`

```
string felipe = "Felipe";
string notFound = "Not Found";
Console.WriteLine(checkIndex(felipe));
Console.WriteLine(checkIndex(notFound));


static string checkIndex(string name)
{
    List<string> names = ["<name>", "Ana", "Felipe"];
    int index = names.IndexOf(name);
    if (index >= 0)
    {
        return $"{name} is found";
    } 
    else
    {
        return $"{name} is not found";
    }    
}
```

```
List<string> names = ["<name>", "Ana", "Felipe","Alice"];
BuildList(names);
names.Sort();
BuildList(names);


static void BuildList(List<string> names)
{
    Console.Write('[');
    foreach (string name in names)
    {
        Console.Write($"{name},"); 
    }
    Console.Write(']');
    Console.WriteLine();
}
```

```
int fibNum = 2;

Console.WriteLine(Fib(2));

static int Fib(int position)
{
    if (position == 0) return 0;  // Handle fib(0)
    if (position == 1) return 1;  // Handle fib(1)

    int firstNumber = 0;
    int secondNumber = 1;
    int fibSum = 0;  // Initialize fibSum

    for (int i = 2; i <= position; i++) // Start from 2 since 0 and 1 are handled
    {
        fibSum = firstNumber + secondNumber;
        firstNumber = secondNumber;
        secondNumber = fibSum;
    }

    return fibSum;
}





// 1 + 1 = 2 + 1 = 3 
```

#### Removing whitespace from strings

- `TrimStart()` remove whitespaces in the beginning
- `TrimEnd()` remove whitespaces in the end
- `Trim()` remove whitespaces in the beginning and end

```
string greeting = "      Hello World!       ";
Console.WriteLine($"[{greeting}]");

string trimmedGreeting = greeting.TrimStart();
Console.WriteLine($"[{trimmedGreeting}]");

trimmedGreeting = greeting.TrimEnd();
Console.WriteLine($"[{trimmedGreeting}]");

trimmedGreeting = greeting.Trim();
Console.WriteLine($"[{trimmedGreeting}]");
```

- `string.Replace('what to replace','replace it with')`

```
string replaceMe = "wassup";
string replaced = replaceMe.Replace('s','z');
Console.WriteLine(replaced);
```

- `string.ToLower()`
- `string.ToUpper()`

```
string pcap = "pcap";
Console.WriteLine(pcap.ToUpper());
Console.WriteLine(pcap.ToLower());
```

- using `Contains` with string

```
string usingContains = "do you see it?";
bool seeIt = usingContains.Contains('d');
Console.WriteLine(seeIt);
```
